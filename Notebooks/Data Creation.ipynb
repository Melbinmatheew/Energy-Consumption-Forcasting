```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("..\Dataset\power-laws-forecasting-energy-consumption-training-data.csv",sep=';')
test=pd.read_csv("..\Dataset\power-laws-forecasting-energy-consumption-test-data.csv",sep=';')

print("Train:",train.head())
print()
print("Test :", test.head())
```

```python
df = pd.concat([train, test], ignore_index=True)
print("Shape:",df.shape)
df
```

```python
value_count=df.groupby('SiteId')['Value'].count()
top_10=value_count.sort_values(ascending=False)[:10]
top_10
```

```python
top_10_sites = value_count.sort_values(ascending=False).head(10).index
```

```python
new_df = df[df['SiteId'].isin(top_10_sites)]
new_df
```

```python
weather_data=pd.read_csv(r"..\Dataset\power-laws-forecasting-energy-consumption-weather.csv",sep=';')
weather_data
```

```python
weather_data['SiteId'].unique()
```

```python
# Merge the DataFrames
merged_df = pd.merge(df, weather_data, on=['SiteId', 'Timestamp'], how='inner')
merged_df
```

```python
Holiday_data = pd.read_csv("..\Dataset\power-laws-forecasting-energy-consumption-holidays.csv", delimiter=';')
Holiday_data
```

```python
# Merge the DataFrames
final_df = pd.merge(merged_df, Holiday_data, on=['SiteId'], how='left')
final_df
```
